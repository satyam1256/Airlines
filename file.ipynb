```python
tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", conn)
tables
```

### `Data Exploration`

```python
aircrafts_data = pd.read_sql_query("select * from aircrafts_data", conn)
aircrafts_data
```

```python
data = aircrafts_data['model']


# Wrapping each English name inside a separate JSON object
english_json_objects = [ data['en'] for data in aircrafts_data['model']]

# Creating a Pandas Series from the JSON objects
series = pd.Series(english_json_objects)

# Replacing the 'model' key in aircrafts_data with the Pandas Series
aircrafts_data['model'] = series

print(aircrafts_data)
```

```python
airports_data = pd.read_sql_query("select * from airports_data", conn)



data = airports_data['airport_name']
english_names = [json.loads(data)['en'] for data in airports_data['airport_name']]
# series = pd.Series(english_names)
airports_data['airport_name'] = english_names
# print(airports_data)


data1 = airports_data['city']
english_names1 = [json.loads(data1)['en'] for data1 in airports_data['city']]
# series1 = pd.Series(english_names1)

airports_data['city'] = english_names1



airports_data
```
